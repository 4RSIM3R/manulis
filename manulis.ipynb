```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import textwrap
```

```python
motivasi = "karena pekerjaannya yg fleksibel dan banyaknya peluang kerja yg ada di jurusan ini, selain itu, saya ingin membuat sistem/aplikasi yg nantinya berguna untuk masa kedepannya"

wrap_motivasi = textwrap.fill(motivasi, 65)
print(wrap_motivasi)
```

```python
folio = Image.open('folio.jpeg')
plt.imshow(folio)
```

```python
tulisan = Image.new('RGBA', (450, 300), (255, 0, 0, 0))
tulisan_draw = ImageDraw.Draw(tulisan)
font = ImageFont.truetype("ShadowsIntoLight-Regular.ttf", 18)

split_motivasi = wrap_motivasi.splitlines()

y_motivasi = 0

for single_motivasi in split_motivasi:
    y_motivasi += 20
    tulisan_draw.text((10, y_motivasi), single_motivasi, "#000000", font=font)

tulisan.save('tulisan.png')

plt.imshow(tulisan)
```

```python
folio_copy  = folio.copy()
tulisan1 = Image.open('tulisan.png')

folio_copy.paste(tulisan1, (150, 250), tulisan1)

folio_copy.save('folio_copy.png')

plt.imshow(folio_copy)
```
